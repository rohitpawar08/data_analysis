**Q: What is the total number of corona cases all over the world (WHO Region wise) ?**

```python
total_number_of_corona_cases = world_corona_virus_cases.groupby('WHO Region')[['Confirmed']].sum()
```

```python
total_number_of_corona_cases
```

```python
plt.figure(figsize=(12,6))
plt.title('Confirmed Cases')
plt.pie(total_number_of_corona_cases, labels=total_number_of_corona_cases.index, autopct='%1.1f%%', startangle=180);
```

**Q: What is the total number of corona deaths all over the world (WHO Region wise) ?**
